### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3
````

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
````

````python
import pandas as pd
import matplotlib.pyplot as plt  

company_sales = pd.read_csv("data/company_sales_data.csv")
monthList  = company_sales ['month_number'].tolist()
faceCremSalesData   = company_sales ['facecream'].tolist()
faceWashSalesData   = company_sales ['facewash'].tolist()
toothPasteSalesData = company_sales ['toothpaste'].tolist()
bathingsoapSalesData   = company_sales ['bathingsoap'].tolist()
shampooSalesData   = company_sales ['shampoo'].tolist()
moisturizerSalesData = company_sales ['moisturizer'].tolist()

plt.plot(monthList, faceCremSalesData, linestyle=':' , label = 'Face cream Sales Data', color='grey', linewidth=2)
plt.plot(monthList, faceWashSalesData,  linestyle='--', label = 'Face Wash Sales Data', color='blue', marker='o', linewidth=2)
plt.plot(monthList, toothPasteSalesData, linestyle='-', label = 'ToothPaste Sales Data', color='red', linewidth=2)
plt.plot(monthList, bathingsoapSalesData, linestyle='-.', label = 'Bathingsoap Sales Data',color='green', linewidth=2)
plt.plot(monthList, shampooSalesData, linestyle='-', label = 'Shampoo Sales Data', color='purple', marker='o', linewidth=2)
plt.plot(monthList, moisturizerSalesData, linestyle='--', label = 'Moisturizer Data', color='blue', marker='^', linewidth=2)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt  
company_sales = pd.read_csv("data/company_sales_data.csv")
monthList  = company_sales ['month_number'].tolist()
toothPasteSalesData = company_sales ['toothpaste'].tolist()
plt.scatter(monthList, toothPasteSalesData, linestyle='-', label = 'ToothPaste Sales Data', color='blue', linewidth=2)
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales data of each month')
plt.show()
````
